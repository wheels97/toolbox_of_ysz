```python
import tensorflow.compat.v1 as tf
import numpy as np
import os
from PIL import Image
```

```python
x = np.array(Image.open(os.path.join(r'C:\Users\85185\tf\inception\img', 'ysz.png')))#读取
resize_0 = tf.image.resize_images(x, (600, 400), method=1)#调整像素格式
#0：双线性差值。1：最近邻居法。2：双三次插值法。3：面积插值法
resize_1 = (resize_0.numpy())
```

```python
import matplotlib.pyplot as plt

plt.figure(0)
plt.imshow(resize_1)
plt.show()
```

```python
#保存图像
import matplotlib.image as mp
mp.imsave('000.png',resize_1)
```
